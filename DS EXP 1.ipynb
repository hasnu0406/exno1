```python
import pandas as pd
import seaborn as sns
```

```python
age=[1,3,28,27,25,92,30,39,40,50,26,24,29,94]
af=pd.DataFrame(age)
af
```

```python
sns.boxplot(data=af)
```

```python
q1=af.quantile(0.25)
q2=af.quantile(0.5)
q3=af.quantile(0.75)
iqr=q3-q1
iqr
```

```python
lb=q1-1.5*iqr
ub=q3+1.5*iqr
```

```python
lb
```

```python
ub
```

```python
af=af[((af>=lb)&(af<=ub))]
```

```python
af
```

```python
from scipy import stats
import numpy as np
data={'weight':[12,15,18,21,24,27,30,33,36,39,42,45,48,51,54,57,60,63,66,69,202,72,75,78,81,84,232,87,90,93,96,99,258]}
df=pd.DataFrame(data)
df 
```

```python
z=np.abs(stats.zscore(df))
print(df[z['weight']>3])
```

```python
val=[12,15,18,21,24,27,30,33,36,39,42,45,48,51,54,57,60,63,66,69,202,72,75,78,81,84,232,87,90,93,96,258]

out=[]
def d_o(val):
    ts=3
    m=np.mean(val)
    sd=np.std(val)
    for i in val:
        z=(i-m)/sd
        if np.abs(z)>ts:
            out.append(i)
    return out

op=d_o(val)
op
```
